```python

def trunc(n, digi, limite):

  arr = []
  j = str(n)
  r = j.split(".") #separar o numero em duas partes
  interira = int(r[0]) #parte inteira
  expoente = 0
  num = n
  x = 0 #contador
  

  if(interira > 0): #se a parte inteira for maior que 0
    expoente = len(r[0]) #tamanho da parte inteira
    num = n*(10**-expoente) #tranformando o numero para decimal

  num = str(num)
  tam2 = len(num)
 
  
  while(x < tam2): #colocar cada digito do numero em uma posição
    arr.append(num[x])
    x = x + 1

  tam3 = len(arr) #tamanho da nova lista com os digitos 

  del(arr[2+digi:tam3]) #exclui todos os elementos a partir do elemento determinado pelo quantidade de digitos

  tam3 = len(arr)
  final = ""
  x = 0
  while(x < tam3):
    final = final + arr[x] #concatenar valores da nova lista com os digitos restantes
    x = x+1

  interira = str(interira)

  if(expoente > limite): #caso o expoente ultrapasse o limite determinado
    expoente = expoente - (expoente - limite)

  expoente = str(expoente)
  final = final + " 10^" + expoente
  #concatenar as partes para o resultado final

  print("Numero truncado: ",final)
```

```python





def arredon(n, digi, limite):
  arr = []
  j = str(n)
  r = j.split(".") #separar o numero em duas partes
  interira = int(r[0]) #parte inteira
  expoente = 0
  num = n
  x = 0 #contador


  if(interira > 0): #se a parte inteira for maior que 0
    expoente = len(r[0]) #tamanho da parte inteira
    num = n*(10**-expoente) #tranformando o numero para decimal 
  
  num = str(num)
  tam2 = len(num)
 
  while(x < tam2): #colocar cada digito do numero em uma posição
    arr.append(num[x])
    x = x + 1

  tam3 = len(arr) #tamanho da nova lista com os digitos 

  del(arr[2+digi:tam3]) #exclui todos os elementos a partir do elemento determinado pelo quantidade de digitos

  tam3 = len(arr)
  final = ""
  
  x = 0
  ultimo_dig = 1+digi
  arredondar = int(arr[ultimo_dig]) + 1 #arredondamento do ultimo digito do numero final
  arredondar = str(arredondar)

  while(x < (tam3-1)): #menos o último elemento, que será inserido depois

    final = final + arr[x] #concatenar valores da nova lista com os digitos restantes
    x = x+1
  final = final + arredondar
  interira = str(interira)

  if(expoente > limite): #caso o expoente ultrapasse o limite determinado
    expoente = expoente - (expoente - limite)

  expoente = str(expoente)
  final = final + " 10^" + expoente
  #concatenar as partes para o resultado final

  print("Numero arredondado: ",final)
```

```python
trunc((30.58999999988),5, 5)
arredon((3055355555.55566666),5,6)
```
